```python
# Prime Check

def prime_check(n):
    if n == 1:
        return False 
    
    for i in range(2, n):
        if i * i > n:
            break 
    
        if n % i == 0:
            return False 
        
    return True

prime_check(int(input()))
```

```python
# Farey Sequence

from math import gcd

def farey(n):
    return sorted({
        (p, q) 
        for q in range(1, n + 1) 
        for p in range(q + 1) 
        if gcd(p, q) == 1
    })

def farey_float(n):
    return sorted({
        p/q 
        for q in range(1, n + 1) 
        for p in range(q + 1) 
        if gcd(p, q) == 1}
    )

def farey_string(n):
    frac = sorted({
        (p, q) 
        for q in range(1, n + 1) 
        for p in range(q + 1) 
        if gcd(p, q) == 1
    })
    return [f"{p}/{q}" if q != 1 else str(p) for p, q in frac]

for n in range(4, 7):
    print(f"F_{n} = {farey(n)}")
    print(f"F_{n} (float) = {farey_float(n)}")
    print(f"F_{n} (string) = {farey_string(n)}")
    print()
```

```python
# Matrix Determinant

def det(m):
    n = len(m)
    if n == 1: return float(m[0][0])
    if n == 2: return float(m[0][0] * m[1][1] - m[0][1] * m[1][0])
    return float(
        sum((-1)**j * m[0][j] * det([row[:j] + row[j+1:] for row in m[1:]]) 
        for j in range(n))
    )

test_matrix_2x2 = [[1, 2], [3, 4]]
test_matrix_3x3 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
test_matrix_3x3_non_zero = [[1, 2, 3], [0, 1, 4], [5, 6, 0]]
test_matrix_4x4 = [[1, 0, 2, -1],[3, 0, 0, 5],[2, 1, 4, -3],[1, 0, 5, 0]]

print(det(test_matrix_2x2))
print(det(test_matrix_3x3))
print(det(test_matrix_3x3_non_zero))
print(det(test_matrix_4x4))
```

```python
# Inv Matrix

def inv(m):
    n = len(m)
    A = [row[:] + [0]*n for row in m]
    for i in range(n):
        A[i][n+i] = 1.0
    
    for i in range(n):
        p = max(range(i, n), key=lambda k: abs(A[k][i]))
        if A[p][i] == 0: return None
        A[i], A[p] = A[p], A[i]
        
        d = A[i][i]
        for j in range(2*n):
            A[i][j] /= d
        
        for k in range(n):
            if k != i:
                mul = A[k][i]
                for j in range(2*n):
                    A[k][j] -= mul * A[i][j]
    
    return [[float(A[i][j]) for j in range(n, 2*n)] for i in range(n)]

test_2x2 = [[1, 2], [3, 4]]
test_3x3 = [[6, 1, 1], [4, -2, 5], [2, 8, 7]]
# [[ 0.17647059 -0.00326797 -0.02287582]
#  [ 0.05882353 -0.13071895  0.08496732]
#  [-0.11764706  0.1503268   0.05228758]]
singular = [[1, 2], [2, 4]]

print(inv(test_2x2))
print(inv(test_3x3))
print(inv(singular))
```
